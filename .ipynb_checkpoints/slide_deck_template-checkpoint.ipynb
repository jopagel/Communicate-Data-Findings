## Investigation Overview

> The following analysis ist supposed to analyze relationships between different features of a bikesharing dataset. Here the feature of interest will be the duration of the trips. This analysis will examine the central variables that influence this feature. 

## Dataset Overview

> Ford GoBike System Dateset is a dateset that contains trip data from a bike service for public use. It consists of 239,111 entries and 14 features. These features include variables like, trip duration, start time and end time with date, start station and end station names, start and end coordinates, customer type, year of birth and gender.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from math import sin, cos, sqrt, atan2, radians

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("fordgobike-tripdata.csv")

## data wrangling

# change start- and end-time to datetime format
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# extract hour and date of start- and end-times and assign new features
df['start_hour'] = df['start_time'].dt.hour
df['end_hour'] = df['start_time'].dt.hour  
df['start_hour'] = df['start_time'].dt.hour
df['end_hour'] = df['start_time'].dt.hour
df['date'] =  df['start_time'].dt.date

# calculate distance and assign to new variable
def distance (lat1,lon1,lat2,lon2):
    R = 6373.0
    
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return np.float64(round(1000*distance))

df['distance'] = df.apply(lambda row: distance(row['start_station_latitude'],row['start_station_longitude'],row['end_station_latitude'],row['end_station_longitude']),axis=1)
```

## Distribution of trip Durations

> The first plot introduces the feature of interest, the duration. As you can see its histogram is a right scewed distribution, which reaches it's Maximum at around 400 seconds. That means that in average most of the trips last around that timespan.

```python
size = 50
bins = bins = np.arange(0, df['duration_sec'].max()+size, size)
plt.figure(figsize=[15,7])
plt.hist(df['duration_sec'], bins = bins)
plt.xlim([0,4000])
plt.title('Duration of trips')
plt.xlabel('Duration [sec]')
plt.ylabel('Count');
```

## Duration vs Distance for two different user-types

The next plot presents the relationship between distance and duration of the trips. As the titles below shows, the plot also includes a third variable, the user-type. The left subplot represents trips by Subscribers, wereas the right suplot shows trips by basic Customers. 
Here one can see that there is a certain correlation between the distance and the trip's duration. Next to that there seems to be a difference in distribution for the two user-types. Subscribers tend to have a rather stable trip duration whereas customers show a higher variance in durations for given distances. This phenomenon can be seen by the all-over distribution of the data point in the right plot. In order to verify this finding, we will compare the features Duration and User-type with use of a bivariate plot.

```python
plt.figure(figsize = [15, 10])

plt.subplot(1,2,1)
plt.scatter(data = df[df['user_type'] == 'Subscriber'], x = 'distance', y = 'duration_sec')

plt.xlim([0,9000])
plt.ylim([0,15000])
plt.title('Duration vs Distance of Subscribers')
plt.xlabel('Distance [m]')
plt.ylabel('Duration [sec]')

plt.subplot(1,2,2)
plt.scatter(data = df[df['user_type'] == 'Customer'], x = 'distance', y = 'duration_sec')

plt.xlim([0,9000])
plt.ylim([0,15000])
plt.title('Duration vs Distance of Customers')
plt.xlabel('Distance [m]')
plt.ylabel('Duration [sec]');
```

## Trip Duration for different User-types

As I already mentioned above, this third plot is intended to resume the analysis between the duration of a trip and which type of user sat on the bike. 
This boxplot proves the initial finding of a higher variance in trip duration for basic Customers in comparison to Subscribers. This can be seen in the plot by the interquartile-range of both user-types. Furthermore the median in durations is more than 250 seconds lower, which indicates that trips by Customers last in average longer, than those by the Subscribers.

```python
plt.figure(figsize = [15, 7])
base_color = sb.color_palette()[0]
sb.boxplot(data = df, x = 'user_type', y = 'duration_sec', color = base_color)
plt.ylim([-10, 2000])
plt.title('Trip Duration in different User-types')
plt.xlabel('User Type')
plt.ylabel('Duration [sec]');
```
